## Atributos discretos

```python
numeros = np.random.rand(50)
etiquetas = pd.Series(pd.cut(numeros, 3, labels=['bajo', 'medio', 'alto']))
etiquetas[:5]
```

```python
frecuencias = etiquetas.value_counts()
frecuencias
```

```python
frecuencias.plot(kind='bar')
```
